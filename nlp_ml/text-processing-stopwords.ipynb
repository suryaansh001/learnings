```python
# Import necessary libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

# Download necessary NLTK data files
nltk.download('punkt')
nltk.download('stopwords')

# Sample text to work with
text = """
Stopwords are words which do not contain important meaning and are usually removed from texts.
For example, words like 'is', 'the', 'and' are considered stopwords.
This text can be tokenized into words and sentences, and stopwords will be removed.
"""

# Tokenize the text into sentences
sentences = sent_tokenize(text)

# Tokenize each sentence into words
words = word_tokenize(text)

# Get the list of stopwords in English
stop_words = set(stopwords.words("english"))

# Remove stopwords from tokenized words
filtered_words = [word for word in words if word.lower() not in stop_words]

# Display results
print("Original Sentences:")
for sentence in sentences:
    print(sentence)

print("\nTokenized Words (Before Stopword Removal):")
print(words)

print("\nFiltered Words (After Stopword Removal):")
print(filtered_words)

print("\nTokenized Sentences:")
print(sentences)
```
